```python
# get my weblog from data2 earlier
import pandas as pd

myfile = "bonusData2_weblogs.csv"
df = pd.read_csv(myfile,header = None)
```

```python
# get the http status code column
import re
url = df.iloc[:,3]
reg = 'GET (.+) HTTP/1.1'
httpUrl = []
for test in url:
    r1 = re.findall(reg,test)[0]
    httpUrl.append(r1)
httpUrl[1:10]
```

```python
# draw the date and http status code from the df above
df.date = df.iloc[:,0]
df.date
df.httpStatus =  df.iloc[:,4]
df.httpStatus.head()
```

```python
# simulate count of complaints
import random
complaints_feedbacks = []
for item in df.httpStatus:
    if (item == 200) or (item == 100):
        a = random.randint(0,10)
        complaints_feedbacks.append(a)
    elif (item == 503) or (item == 404):
        a = random.randint(30,100)
        complaints_feedbacks.append(a)
print(complaints_feedbacks)  
        
        
```

```python
# append all the columns to a dataframe
d = {'Datetime': df.date.tolist(),'URL':httpUrl, 'status code':df.httpStatus.tolist(), 'Complaints Or Feedbacks':complaints_feedbacks }
df_All = pd.DataFrame(data=d)
df_All.to_csv("bonusData3_Complaints_Feedbacks.csv")
df_All
```
