```python
# draw out the unique pool of usernames from ordersdata
import pandas as pd
ordersdata = pd.read_csv("ordersdata.csv")
ordersdataCustName =  ordersdata.customername
ordersdataCustName = ordersdataCustName.unique()
len(ordersdataCustName)

#split the names by " "
ordersdataCustNameSplit = [row.split(" ") for row in ordersdataCustName]
#print(ordersdataCustNameSplit)

# join the names by "_"
ordersdataCustNameJoin = ["_".join(row1) for row1 in ordersdataCustNameSplit]
print(ordersdataCustNameJoin)
```

```python
# create a list of running numbers to append the IP add, then later random pick the IP
index = list(range(len(ordersdataCustName)))
```

```python
# make a list of 100 different ip address
# col 1 ip
# realistically, the digits of IP run from 0 to 255, but we choose 50 to 250 since some numbers are rare
import random
IP = []
for x in range(len(ordersdataCustName)):
    a = random.randint(50,250)
    b = random.randint(50,250)
    c = random.randint(50,250)
    d = random.randint(50,250)
    ip_add = "{}.{}.{}.{}".format(a,b,c,d)
    IP.append(ip_add)
IP[1:10]
```

```python
# start off a new dataframe by creating from these 2 columns above
import pandas as pd
import numpy as np
d = {'ind': index, 'IP': IP,'CustName':ordersdataCustNameJoin}
df = pd.DataFrame(data=d)
df.head()
len(df)
df.head()
```

```python
# randomly choose an ind to get the IP at random
import random
def genId(rows):
    randomIdx = []
    for x in range(rows):
        randId = random.randint(0,len(df)-1)
        randomIdx.append(randId)
    return(randomIdx)
gen = genId(20)
print(gen)
```

```python
# take the corresponding column of IP addresses. This method ensures it is randomly selected
# also ensures IP can be retaken again
# combine string IP and ordersdataCustName
mergedCol = []
for row in range(0,len(df)):
    merged = '{},{}'.format(df.IP[row], df.CustName[row])
    mergedCol.append(merged)
#print(mergedCol)

# genIP based on the random num for row
genIP = []
for i in gen:
    genIP.append(mergedCol[i])
genIP
```

```python
# number of times each IP is repeated. 
# create a random list of numbers to repeat
import random
import numpy as np

repeatTimes = []
for repeat in range(len(genIP)):
    rep = random.randint(1,10) # browsing 1 to 10 times of a website is reasonable
    repeatTimes.append(rep)
print('repeatTimes:{}'.format(repeatTimes))

# # create a list of repeated IP addresses
# x = np.array(genIP)
# print(np.repeat(x, repeatTimes, axis=0))
```

```python
# create realistic time jump
# between users time jump bigger
# for each user at one session, timejump not big
# make url address
# ignore column of referrer
# there are 3 forms 
# first form : the home page "GET / HTTP/1.1"
# 2nd form : "GET /products/{} HTTP/1.1" where {} is a number
# 3rd form : "GET /products/images/{}.jpg HTTP/1.1" where this num 
# same as above, since this website likely got pics every page (consumer buy things)
# customers use "GET"
# there is a weighted selection of HTTP codes as there should be less breakdown than OK mode 
# https://stackoverflow.com/questions/3679694/a-weighted-version-of-random-choice
from datetime import timezone
import datetime
import pytz

#https://janakiev.com/blog/time-and-timezones-in-python/
def getfilesize(prodId,filetype):
    if filetype == 'image':
        filesize = (prodId)*11 + 200
    elif filetype == 'noimage':
        filesize = (prodId)*5 + 200
    else:
        print("error")
    return(filesize)

# initial time is 01-01-2019
x = datetime.datetime(2019, 1, 1,0,0,0) 
x = pytz.timezone('Asia/Singapore').localize(x)
#print(x)
counterRep = 1
# start = datetime.datetime(2019, 1, 1,0,0,0) 
end = datetime.datetime(2019, 3, 31,23,59,59)
end = pytz.timezone('Asia/Singapore').localize(end)

# https://stackoverflow.com/questions/3679694/a-weighted-version-of-random-choice
# later need to random choice by weight
http = [503,404,200,100] # 503 and 404 are errors which user will see. 200 and 100 are OK signs


def numRepeatGen(initialTime, writer):
    counter1 = 0
    
    
    for ip in genIP:
        initialTime = initialTime + datetime.timedelta(seconds=random.randint(70,300*60))
        rep = random.randint(1,10) # browsing 1 to 10 times of a website is reasonable
        for i in range(rep):
            counter1 = counter1+1
            #print(ip, i)
            initialTime = initialTime + datetime.timedelta(seconds=random.randint(1,5*60))
            # get url
            if initialTime <= end:
                if i == 0 :
                    url = "'GET / HTTP/1.1'"
                    filesize = 199
                    print(initialTime, ip,url, random.choice(http), filesize)
                    toprint = '{},{},{},{},{}\n'.format(initialTime,ip,url,random.choices(
population=http,weights=[0.05,0.05,0.70,0.20],k=1)[0],filesize)
                    writer.write(toprint)
                else : 
                    randProdId = random.randint(1,51)
                    url1 = "'GET /products/{} HTTP/1.1'".format(randProdId)
                    url2 = "'GET /products/images/{}.jpg HTTP/1.1'".format(randProdId)   
                    filesize1 = getfilesize(randProdId,'noimage')
                    filesize2 = getfilesize(randProdId,'image')
                    print(initialTime, ip, url1,random.choice(http), filesize1)
                    toprint = '{},{},{},{},{}\n'.format(initialTime,ip,url1,random.choices(
population=http,weights=[0.05,0.05,0.70,0.20],k=1)[0],filesize1)
                    writer.write(toprint)
                    print(initialTime, ip, url2,random.choice(http), filesize2)
                    toprint = '{},{},{},{},{}\n'.format(initialTime,ip,url2,random.choices(
population=http,weights=[0.05,0.05,0.70,0.20],k=1)[0], filesize2)
                    writer.write(toprint)
            else:
                print("reached date limit")
                return(initialTime)
    print(counter1)
        
        
    return(initialTime)
```

```python
with open('bonusData2_weblogs.csv', 'w+') as writer:
    endTime = numRepeatGen(x, writer)  # x from above, where it is 01-01-2019
    counter =1
    while endTime < end:
        endTime = numRepeatGen(endTime, writer) # writer for func to write to this file
        counter = counter +1
        #print(counter)

```
